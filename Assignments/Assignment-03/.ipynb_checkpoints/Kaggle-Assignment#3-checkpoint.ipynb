## Spliting data into taining and Testing Set

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

## Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test,model.predict(X_test))

print("Mean Squred Error :\t" , math.sqrt(mse))
print("R Squared Value is :\t", model.score(X_train,y_train))
```

## Comperision with svm

```python
dataset = pd.read_csv('insurance.csv')
dataset.head(10)
```

```python
dataset.columns
```

```python
dataset[['smoker']] = dataset[['smoker']].replace('no' , 0)
dataset[['smoker']] = dataset[['smoker']].replace('yes' , 1)
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
X =  dataset[['bmi']]
y = dataset[['charges']]
print(X)
print(y)
```

```python
X = np.array(X).reshape(-1,1)
y = np.array(y).reshape(-1,1)
```

```python
X
```

```python
model = SVR(kernel = 'rbf')
model.fit(X,y)
```

```python
y_pred = model.predict(X)
y_pred
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = "purple")
plt.plot(X_grid, model.predict(X_grid), color = "black")
plt.title("Truth or Bluff (Support Vector Regression Model(High Resolution))")
plt.xlabel("Body Mass Index")
plt.ylabel("charges")
plt.show()
```

```python
mse = mean_squared_error(y,model.predict(X))

print("Mean Squred Error :\t" , math.sqrt(mse))
print("R Squared Value is :\t", model.score(X,y))
```
