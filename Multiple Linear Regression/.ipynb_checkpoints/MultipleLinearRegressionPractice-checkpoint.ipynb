```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error
import math
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
df.head(10)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df=df.drop(['id' , 'date','waterfront' , 'view'] , axis=1)
```

```python
df.head(10)
```

```python
sns.pairplot(df[['sqft_lot','sqft_above','price','sqft_living','bedrooms']], hue='bedrooms', palette='tab20',size=6)
```

```python
X = df.iloc[:,1:].values#[rows,columns] columns ki sari rows or pehla column chor k baki sary column
print(X)
y = df.iloc[:,0].values # first column of sari rows
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred.sum().mean())
print(y_test.sum().mean())

print("So called Accuracy ", y_pred.sum().mean()/y_test.sum().mean())
```
